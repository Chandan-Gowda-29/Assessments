# Data preprocessing

```python
data = data_import.copy()
data.rename(columns={'Unnamed: 0': 'Customer Query'}, inplace=True)
```

```python
data
```

```python
# Dropping unwanted columns
data.drop(data.iloc[:, 13:27], inplace = True, axis = 1)
```

```python
# Splitting of string in "Costomer Query" column to extract features
data = pd.concat([data["Customer Query"].str.split('"',expand=True),data],axis=1)
```

```python
columns_to_discard = [feature for feature in data.columns if len(data[feature].unique())<2]
```

```python
data.drop(data[columns_to_discard], inplace = True, axis = 1)
```

```python
data = data.drop(['Customer Query'], axis = 1)
```

## Renaming the newly added columns in the data 

```python
#Renaming of features taken fron "Customer Query"
data.rename(columns={7:'Cq_givenName',11:'Cq_familyName',17:'Cq_address',21:'Cq_city',25:'Cq_regionCode',28:'Cq_postalCode',33:'Cq_emails',39:'Cq_number'}, inplace=True)
```

```python
import re
l=[]
for i in range(0, len(data['Cq_postalCode'])):
    nn = re.sub("[^0-9]", "", data['Cq_postalCode'][i])
    l.append(nn)
```

```python
data["Cq_postalCode"] = l
```

```python
data
```

## Replacing the empty white cells(missing values) with 0 

```python
# Replacing all the missing values with NaN
data.replace('', 0, inplace=True)
data
```

### Number of missing values in each row 

```python
m = (data == 0).astype(int).sum(axis=1)
m
```

```python
data['Missing values'] = m
data
```

```python
plt.plot(data['Confidence'])
plt.plot(data['Missing values'])
```

The number of missing values in case of each person doesn't seem to be effecting the Confidence.

## Comparing of the features from  'Customer Query' with corresponding features delivered by the company

```python
data['Cq_postalCode']=data['Cq_postalCode'].astype(str).astype(int)
```

```python
data['FIRST NAME'] = data.apply(lambda data: data['Cq_givenName'] == data['first_name'], axis=1)
data['LAST NAME'] = data.apply(lambda data: data['Cq_familyName'] == data['last_name'], axis=1)
data['ADDRESS'] = data.apply(lambda data: data['Cq_address'] == data['address'], axis=1)
data['CITY'] = data.apply(lambda data: data['Cq_city'] == data['city'], axis=1)
data['STATE'] = data.apply(lambda data: data['Cq_regionCode'] == data['state'], axis=1)
data['ZIP'] = data.apply(lambda data: data['Cq_postalCode'] == data['zip'], axis=1)
data['EMAIL'] = data.apply(lambda data: data['Cq_emails'] == data['email'], axis=1)
data['NUMBER'] = data.apply(lambda data: data['Cq_number'] == data['phone1'], axis=1)

data
```

```python
data['Total matching features'] = data[['FIRST NAME', 'LAST NAME', 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'EMAIL', 'NUMBER']].sum(1)
data
```

```python
data['Confidence'].corr(data['Total matching features'])
```

Total number of matching values is not effecting the Confidence values.

## Confidence values doesn't seem to be capturing the effect of matching and missing data. Hence we can calculate an alternate confidence value as shown below

```python
# Calculate Percentage based on matching values of input and output
data['CONFIDENCE'] = data['Total matching features'] /8
data
```

```python
var = ['Confidence', 'CONFIDENCE']
```

```python
fig=plt.subplots(figsize=(12,8))
plt.plot(data['Confidence'],label='Original')
plt.plot(data['CONFIDENCE'],label='Derived')
plt.legend(fontsize='x-large')
plt.title('Original and derived Confidence', size=20)
plt.xlabel('Row Numbers',size=15)
plt.ylabel('Confidence',size=15)
plt.show()
```
